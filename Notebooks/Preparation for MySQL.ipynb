```python
# Selecting columns for the interventions table
df_interv = df[['study_id', 'Interventions']].copy()

# Separating type from details
df_interv['Interventions'] = df_interv['Interventions'].str.split('|')
df_interv = df_interv.explode('Interventions')

# Checking for missing values in date columns
df_interv[['Interventions']].isna().sum()


```

```python
# Removing empty rows
df_interv = df_interv.dropna(subset=['Interventions'])

# Dividing the 'Interventions' column into 'type' and 'description'
df_interv[['intervention_type', 'description']] = df_interv['Interventions'].str.split(':', n=1, expand=True)

# Stripping whitespace
df_interv['intervention_type'] = df_interv['intervention_type'].str.strip()
df_interv['description'] = df_interv['description'].str.strip()

# Selecting final Interventions table
df_interventions = df_interv[['study_id', 'intervention_type', 'description']] # already in the correct name format

df_interventions
```

```python
# Exporting
df_interventions.to_csv(os.path.join(export_path, 'interventions.csv'), index=False)
```

```python
# Selecting columns for the sponsors table
df_sponsors_raw = df[['study_id', 'Sponsor/Collaborators', 'Funded Bys']].copy()

# Renaming columns to match the database schema
df_sponsors_raw = df_sponsors_raw.rename(columns={'Funded Bys': 'agency_class'})

# Function to separate agencies and define who is Lead or Collaborator
# The first agency in the list is considered the Lead, and the others are Collaborators
def process_sponsors(row):
    '''
    This function takes a row of the dataframe, separates the sponsors/collaborators and their classes,
    and returns a list of dictionaries with the agency name, class, and role (Lead or Collaborator).
    '''
    
    agencies = str(row['Sponsor/Collaborators']).split('|')
    classes = str(row['agency_class']).split('|')
    
    results = []
        
    for i, name in enumerate(agencies):
        # The first in the list (index 0) is the Lead, the others are Collaborator
        role = 'Lead' if i == 0 else 'Collaborator'
        
        # Find the corresponding class (if it exists in the same order)
        # If no class is found for all, mark as 'Other' or 'Unknown'
        current_class = classes[i].strip() if i < len(classes) else "Other"
        
        results.append({
            'agency': name.strip(),
            'agency_class': current_class,
            'lead_or_collaborator': role
        })
    return results

# Creating a temporary column with the list of dictionaries
df_sponsors_raw['temp_list'] = df_sponsors_raw.apply(process_sponsors, axis=1)

# Separating the list of dictionaries into individual rows
df_sponsors_final = df_sponsors_raw.explode('temp_list')

# Converting the list of dictionaries into separate columns and cleaning the dataframe
df_sponsors_final = pd.concat([
    df_sponsors_final[['study_id']], 
    df_sponsors_final['temp_list'].apply(pd.Series)
], axis=1)

# Final cleaning
df_sponsors_final = df_sponsors_final[df_sponsors_final['agency'] != 'nan']
df_sponsors_final = df_sponsors_final.dropna(subset=['agency'])

df_sponsors_final.head(10)
```
